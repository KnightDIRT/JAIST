## Create a voice clone  
Chatterbox will generate a voice embedding from your sample.

```python
from chatterbox.tts import ChatterboxTTS
import torch

# Set device: "cuda" if GPU available, otherwise "cpu"
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)

# Load model on the selected device
model = ChatterboxTTS.from_pretrained(device=device)

model.prepare_conditionals(audio_path)
    
print("Model created successfully!")
```

## Generate TTS using the cloned voice


```python
import torchaudio as ta

text = "Crazy? I Was Crazy Once. They Locked Me In A Room. A Rubber Room. A Rubber Room With Rats. And Rats Make Me Crazy."

# Generate audio using your cloned voice
wav = model.generate(text,
                     exaggeration=0.5, # 0 monotone, 0.5 default, >=1 exaggerated
                     temperature=0.8, # randomness: 0.8 default, higher more random
                     top_p=1.0, # probability distributiom: 0.5 controlled, 1.0 default fully open
                     min_p=0.05, # lower more expressive but more noisy, 0.05 default
                     repetition_penalty=1.2, # more variety but may distort voice, 1.2 default
                     cfg_weight=0.5 # guidance scale: 0 no guidance, 0.5 default, higher more guided clear but emotionless
                     )

output_path = "tts_output.wav"

# Save the generated audio to a file
ta.save(output_path, wav, model.sr)

print("Saved:", output_path)
```
